```python
import pandas as pd
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
df = pd.read_csv("/Users/iansmeenk/Voodoo/plays_2017.csv")
```

```python
df.head()
```

```python
df2 =df.query("gsis_game_id == 66749")
```

```python
df2.head()
```

```python
cols = ['play_index','time_in_reg', 'offteam', 'defteam', 'field_position']
```

```python
df3 = df2.loc[:, cols]
```

```python
game = df3.sort_values('time_in_reg')
```

```python
game.head()
```

```python
def flip_fp(row):
    if row["defteam"]=='CAST':
        row["field_position"] = 100 - row["field_position"]
    return row
```

```python
game2 =game.apply(flip_fp, axis=1)
```

```python
trace0 = go.Scatter(
    x=game2['time_in_reg'],
    y=game2['field_position'],
    mode='lines',
    line=dict(width=0.5,
              color='rgb(189, 32, 49)'),
    fill='tonexty'
)
trace1 = go.Scatter(
    x=game2['time_in_reg'],
    y=[100]*len(game2['time_in_reg']),
    mode='lines',
    line=dict(width=0.5,
              color='rgb(0, 121, 53)'),
    fill='tonexty'
)

data = [trace0, trace1]
layout = go.Layout(
    showlegend=True,
    xaxis=dict(
        type='category',
        dtick=20
    ),
    yaxis=dict(
        type='linear',
        range=[1, 100],
        dtick=20
    )
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='stacked-area-plot')
```
